```python
import pandas as pd
import datetime as dt
import matplotlib as m
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv("paysim.csv")
```

```python
data.head()
```

```python
node=data.loc[data['isFraud'] == 1]  
```

```python
import matplotlib.pyplot as plt
import networkx as nx
node_x=node[['nameOrig','nameDest','amount']][1:20]
#node_x
H = nx.from_pandas_edgelist(node_x, 'nameOrig', 'nameDest','amount')
G = nx.DiGraph()
G.add_edges_from(H.edges())
pos = nx.random_layout(G)
nx.draw_networkx_nodes(G, pos, cmap=plt.get_cmap('jet'), 
                        node_size = 500)
nx.draw_networkx_labels(G, pos)
nx.draw(G,pos,node_size=100,font_size=100)
#nx.draw_networkx_edges(G, pos,  edge_color='r', arrows=True)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(),edge_color='b',arrows=True)
plt.figure(3,figsize=(100,100)) 
plt.show()
```
