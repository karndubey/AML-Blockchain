```python
import pandas as pd
import datetime as dt
import matplotlib as m
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv("dataset.csv")
data['operation'] = data['operation'].replace(['CASH IN'], 0)
data['operation'] = data['operation'].replace(['CASH WITHDRAWAL'], 1)
data['operation'] = data['operation'].replace(['BANK TRANSFER'], 2)
df=data.loc[data['account_id'] == 9]
mylist=list(df['date1']) 
l=[dt.datetime.strptime(d,'%m/%y').date() for d in mylist] 
dates = m.dates.date2num(l) 
plt.figure(figsize=(20,10))
plt.xlabel("Date",fontsize=20)
plt.ylabel("Amount",fontsize=20)
plt.suptitle("Time Series Plot",fontsize=30)
plt.plot_date(dates, df['amount'],'-')
#plt.figure(figsize=(20,20))
#plt.plot(df['date'],df["amount"])
```

```python
arr = np.array(df['amount'])
x=float(sum(arr)/len(df['amount']))
sigma=np.std(arr)
k=np.array(df)
print(x)
print(sigma)
for i in range(len(k)):
    if(abs((k[i][4]))-x>2*sigma):
        print('amount ',k[i][4])
        print('operation',k[i][3])
        print('date' ,k[i][8])
        print((k[i][4]-x)/k[i][4])
        print("\n")
```

## frequency plot with 3 dim..

```python
k=0
list1 =[]
list2 = []
list3 = []
list1.append(0)
list2.append(0)
list3.append(0)
if(df['operation'][0]==0):
        list1[k]+=1
elif(df['operation'][0]==1):
        list2[k]+=1;
else:
        list3[k]+=1
for i in range(1,len(df)):
    if(df['date1'][i]!=df['date1'][i-1]):
        k=k+1
        list1.append(0)
        list2.append(0)
        list3.append(0)
    if(df['operation'][i]==0):
        list1[k]+=1
    elif(df['operation'][i]==1):
        list2[k]+=1;
    else:
        list3[k]+=1
```

```python
def func(items):
    seen=set()
    for item in items:
        if(item not in seen):
            yield item
            seen.add(item)
    return seen
var=list(func(dates))
plt.plot_date(var,list1,'o')
```
